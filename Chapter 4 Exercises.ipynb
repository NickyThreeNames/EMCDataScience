Using Ruspini dataset, perform k-means evaluation

```r
library(cluster)
data(ruspini)

#Decide on number of clusters
r <- ruspini
wss <- numeric(15)

for( k in 1:15) wss[k] <- sum(kmeans(r, centers = k, nstart = 25)$withinss)
    
plot(1:15, wss, type = 'b', xlab = "Num clusters", ylab="Within Sum of Squares")
```

```r
#Looking at the plot we'll use four clusters.

km = kmeans(r, 4, nstart = 25)
km
```

```r
#evaluate clusters
library(ggplot2)
df = as.data.frame(r)
df$cluster = factor(km$cluster)
df
ggplot(data = df, aes(x = x, y = y, color = cluster))+
```
